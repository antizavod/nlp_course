```python
# surname recognition version 0.1
# surnames such as Rubtsov(a), Andreev(a), Polonskiy
Word = u'Жаботински'
def Surnames(Word):
#if Word != Word.lower():
    try:
        end = re.search(u'ов|ев|ёв|ск', Word[-5:]).group(0)
        Surname = re.match(u'^[А-Я].*' + end, Word).group(0)
        if end == u'ск':
            Surname += u'ий'
    except:
        Surname = Word
    return Surname
```
